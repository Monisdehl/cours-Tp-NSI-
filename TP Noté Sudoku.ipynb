### 8) Vous venez de faire une grille de sudoku réponse. Maintenant faisons la grille question. Tester plusieurs fois la fonction suivante avec plusieurs valeurs de $p$ compris entre 0 et 1. Que comprenez vous

```python
import random as rd
def random_hide(p,k):
    if rd.random()<p:
        return ""
    else:
        return k
    
p = 0.75
print(random_hide(p,3))
    
```

la fonction random_hide(p,k) prend un nombre au hasard entre 0 et 1(exclu). Si sa valeur est inférieure à celle de p (entre 0 et 1 inclu(s)) alors rien n'apparait. Sinon k apparait. Cela correspond au pourcentage de chance que k n'apparaisse pas (ici 75%).
